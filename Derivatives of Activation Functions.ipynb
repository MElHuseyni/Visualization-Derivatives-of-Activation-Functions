# Derivatives of Activation Functions

```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline 
# https://www.youtube.com/watch?v=P7_jFxTtJEo
```

## Sigmoid Function

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.,5.,0.01)
y = sigmoid(x)
dx = y*(1-y)
```

```python
fig,ax = plt.subplots(figsize=(9,5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x,y,color='b',linewidth=3,label='Sigmoid')
ax.plot(x,dx,color='r',linewidth=3,label='Derivative')
ax.legend(loc="upper left",frameon=True)
plt.grid();
```

## Tanh Fucntion (Tangent Hyperbolic)

```python
tanh = lambda z:((np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z)))
```

```python
z = np.arange(-4.,4.,0.01)
a = tanh(z)
dz = 1 - a**2
```

```python
fig,ax = plt.subplots(figsize=(9,5))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.plot(z,a,color='b',linewidth=3,label='Tanh')
ax.plot(z,dz,color='r',linewidth=3,label='Derivative')
ax.legend(loc="upper left",frameon=True)
plt.grid();
```

## ReLU Fucntion (Rectified Linear Unit)

```python
ReLU = lambda x: x * (x > 0)
dReLU =  lambda x: 1. *(x > 0)
```

```python
x = np.arange(-6.,6.,0.01)
y = ReLU(x)
dx = dReLU(x)
```

```python
fig,ax = plt.subplots(figsize=(9,5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.plot(x,y,color='b',linewidth=3,label='ReLU')
ax.plot(x,dx,color='r',linewidth=3,label='Derivative')
ax.legend(loc="upper left",frameon=True)
plt.grid();
```
